# 1. Correlacion Pearson (Sin Usar Kmeans)

```python
import matplotlib.pyplot as plt
import numpy as np
from scipy.stats import pearsonr
```

```python
x = [185, 170, 168, 179, 182, 188]
y = [72, 56, 60, 68, 72, 77]
```

```python
# Calcular la correlación de Pearson
correlacion, _ = pearsonr(x, y)
print("Correlación de Pearson:", correlacion)
```

```python
# Crear un gráfico de puntos con leyenda
plt.scatter(x, y, label="Datos")
plt.title('Gráfico de Puntos')
plt.xlabel('Coordenada X')
plt.ylabel('Coordenada Y')
plt.legend()
plt.grid(True)
```

```python
# Crear una matriz de correlación
data = np.array([x, y])
correlacion_matrix = np.corrcoef(data)
plt.subplot(1, 2, 2)
plt.imshow(correlacion_matrix, cmap='viridis', origin='lower', aspect='auto')
plt.colorbar()
plt.title('Matriz de Correlación (sin K-Means)')
plt.xticks([0, 1], ['X', 'Y'])
plt.yticks([0, 1], ['X', 'Y'])

plt.show()
```

# 2. Correlacion Pearson (Usando Kmeans)

```python
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
import numpy as np
from scipy.stats import pearsonr
```

```python
x = [185, 170, 168, 179, 182, 188]
y = [72, 56, 60, 68, 72, 77]
```

```python
# Combinar las listas de coordenadas en un solo conjunto de datos
data = np.array(list(zip(x, y)))
```

```python
# Número de clústeres
num_clusters = 2

# Crear un modelo K-Means
kmeans = KMeans(n_clusters=num_clusters, random_state=0)

# Ajustar el modelo a los datos
kmeans.fit(data)

# Obtener las etiquetas de clúster asignadas a cada punto
labels = kmeans.labels_

# Obtener los centroides
centroides = kmeans.cluster_centers_
```

```python
# Crear un gráfico de puntos
plt.figure(figsize=(12, 4))
plt.subplot(1, 2, 1)
plt.scatter(x, y, c=labels, cmap='viridis', label="Datos")
plt.scatter(centroides[:, 0], centroides[:, 1], marker='X', s=100, color='red', label='Centroides')
plt.title('Gráfico de Puntos con K-Means')
plt.xlabel('Coordenada X')
plt.ylabel('Coordenada Y')
plt.legend()
```

```python
# Crear una matriz de correlación
correlacion_matrix = np.corrcoef(data.T)
plt.subplot(1, 2, 2)
plt.imshow(correlacion_matrix, cmap='viridis', origin='lower', aspect='auto')
plt.colorbar()
plt.title('Matriz de Correlación con K-Means')
plt.xticks([0, 1], ['X', 'Y'])
plt.yticks([0, 1], ['X', 'Y'])

plt.show()
```

```python
# Calcular la correlación de Pearson
correlacion, _ = pearsonr(x, y)
print("Correlación de Pearson (sin K-Means):", correlacion)

correlacion, _ = pearsonr(centroides[:, 0], centroides[:, 1])
print("Correlación de Pearson (con K-Means):", correlacion)
```
